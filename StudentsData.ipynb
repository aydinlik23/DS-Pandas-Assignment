```python
import numpy as np
import pandas as pd

FB = pd.ExcelFile("D:\Python Course\Data Science\Week-2\Assignment\FB.xlsx")
GS = pd.ExcelFile("D:\Python Course\Data Science\Week-2\Assignment\GS.xlsx")
BJK = pd.ExcelFile("D:\Python Course\Data Science\Week-2\Assignment\BJK.xlsx")
```

```python
files = [FB, GS, BJK]
classes=[]
names=[]
correct=[]
incorrect=[]
blank=[]

for i in files:                                     
    for j in range(len(i.sheet_names)):                
        
        Datas = pd.read_excel(i, i.sheet_names[j]) 
        
        names += [Datas.columns[0]]                
        correct += [Datas.iloc[-3,1]]
        incorrect += [Datas.iloc[-2,1]]
        blank += [Datas.iloc[-1,1]]
        
        if i == FB:
            Class = "FB"
        elif i == GS:
            Class = "GS"
        elif i == BJK:
            Class = "BJK"
       
        classes += [Class]

all_students = pd.DataFrame({"Class": classes, "Students-Name": names ,"Correct": correct, 
                             "Incorrect": incorrect, "Blank": blank})

all_students.index = all_students.index + 1     
all_students
```
